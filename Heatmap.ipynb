```python
def f(X, Y, f00, f05, f10):
    """
    Двухступенчатая интерполяция функции вероятности окружения f(X, Y):
    X - число дивизий пехоты
    Y - превосходство в воздухе у пехоты (0 - зеленый воздух, 0.5 - желтый, 1 - красный)
    Для нахождения значения в произвольной точке (X, Y) имея сетку при X = 0, 1, ... 50 и Y = 0, 0.5, 1
    - сначала идет линейная интерполяция по Y (0, 0.5, 1)
    - потом линейная по X
    """

    x1 = min(math.floor(X), len(f00) - 2)
    x2 = x1 + 1

    # ЛИНЕЙНАЯ интерполяция по Y (вместо Лагранжа, чтобы сохранить монотонность)
    if Y <= 0.5:
        tY = Y / 0.5
        L0 = 1 - tY
        L05 = tY
        L1 = 0.0
    else:
        tY = (Y - 0.5) / 0.5
        L0 = 0.0
        L05 = 1 - tY
        L1 = tY

    # нахождение f(X1, Y)
    f_x1 = (
        f00[x1] * L0 +
        f05[x1] * L05 +
        f10[x1] * L1
    )

    # нахождение f(X2, Y)
    f_x2 = (
        f00[x2] * L0 +
        f05[x2] * L05 +
        f10[x2] * L1
    )

    # нахождение f(X, Y)
    t = X - x1
    return f_x1 * (1 - t) + f_x2 * t


def g(X, Y, g00, g05, g10):
    """
    Двухступенчатая интерполяция функции средней организации на третьей линии g(X, Y):
    X - число дивизий пехоты
    Y - превосходство в воздухе у пехоты (0 - зеленый воздух, 0.5 - желтый, 1 - красный)
    Для нахождения значения в произвольной точке (X, Y) имея сетку при X = 0, 1, ... 50 и Y = 0, 0.5, 1
    - сначала идет линейная интерполяция по Y (0, 0.5, 1)
    - потом линейная по X
    """

    x1 = min(math.floor(X), len(g00) - 2)
    x2 = x1 + 1

    # ЛИНЕЙНАЯ интерполяция по Y (замена Лагранжа для сохранения монотонности)
    if Y <= 0.5:
        tY = Y / 0.5
        L0 = 1 - tY
        L05 = tY
        L1 = 0.0
    else:
        tY = (Y - 0.5) / 0.5
        L0 = 0.0
        L05 = 1 - tY
        L1 = tY

    # нахождение g(X1, Y)
    g_x1 = (
        g00[x1] * L0 +
        g05[x1] * L05 +
        g10[x1] * L1
    )

    # нахождение g(X2, Y)
    g_x2 = (
        g00[x2] * L0 +
        g05[x2] * L05 +
        g10[x2] * L1
    )

    # нахождение g(X, Y)
    t = X - x1
    return g_x1 * (1 - t) + g_x2 * t
```

```python
# сетка
nx, ny = 300, 300
x_vals = np.linspace(0, 30, nx)
y_vals = np.linspace(0, 1, ny)

F_img = np.zeros((ny, nx))
G_img = np.zeros((ny, nx))

for i, y in enumerate(y_vals):
    for j, x in enumerate(x_vals):
        F_img[i, j] = f(x, y, f00, f05, f10)
        G_img[i, j] = g(x, y, g00, g05, g10)

# цвета
A = np.array([1, 0, 0])  # красный, соотвествует 100% маневренной войне (100% вероятность окружения)
B = np.array([0, 1, 0])  # зеленый, соотвествует 100% гибридной войне (100% организация на третьей линии)

# RGB изображение
img = F_img[..., None] * A + G_img[..., None] * B

# отрисовка
width_px = 1600
height_px = 760
dpi = 100

plt.figure(figsize=(width_px / dpi, height_px / dpi), dpi=dpi)
plt.imshow(
    img,
    origin='lower',
    extent=[x_vals.min(), x_vals.max(), y_vals.min(), y_vals.max()],
    aspect='auto'
)
plt.axis('on')
plt.show()
```
